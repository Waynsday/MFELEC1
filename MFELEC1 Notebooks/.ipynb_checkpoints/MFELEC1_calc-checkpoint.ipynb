```python
from tkinter import *      #basic user-interface (UI) of python
```

```python
root = Tk()
root.title("MFELEC1 App") # type = 24     # 15 to 24
root.geometry('300x300')    #window size
root.configure(background = '#087830')   #00FF00 (green)
root.resizable(0,0)

# objects: label, button, entry

def add():
    a = int(text1.get())
    b = int(text2.get())
    c = a + b
    label3 = Label(root, text = c, font = "Times 20 bold italic", bg = '#087830', fg = 'white')
    label3.grid(row = 2, column = 1)
    
def sub():
    a = int(text1.get())
    b = int(text2.get())
    c = a - b
    label3 = Label(root, text = c, font = "Times 20 bold italic", bg = '#087830', fg = 'white')
    label3.grid(row = 3, column = 1)
    
def mult():
    a = int(text1.get())
    b = int(text2.get())
    c = a * b
    label3 = Label(root, text = c, font = "Times 20 bold italic", bg = '#087830', fg = 'white')
    label3.grid(row = 4, column = 1)

label1 = Label(root, text = "1st No:", font = "Times 20 bold italic", bg = '#087830', fg = 'white')
label2 = Label(root, text = "2nd No:", font = "Times 20 bold italic", bg = '#087830', fg = 'white')
button1 = Button(root, text = "+", font = "Arial 20 bold", command = add)  #, bg = '#087830', fg = 'white')
button2 = Button(root, text = "-", font = "Arial 20 bold", command = sub)
button3 = Button(root, text = "x", font = "Arial 20 bold", command = mult)
#button4 = Button(root, text = "Close", font = "Arial 20 bold", command = close)

text1 = Entry(root, w = 20)
text2 = Entry(root, w = 20)

#------------------#  grid vs pack
label1.grid(row = 0, column = 0)
text1.grid(row = 0, column = 1)
label2.grid(row = 1, column = 0)
text2.grid(row = 1, column = 1)
button1.grid(row = 2, column = 0)
button2.grid(row = 3, column = 0)
button3.grid(row = 4, column = 0)
#button4.grid(row = 5)

root.mainloop()
```
