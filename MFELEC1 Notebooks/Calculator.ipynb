```python
from tkinter import *
from functools import partial
```

```python
root = Tk()
root.title("MFELEC1 Calculator") # type = 24     # 15 to 24
root.geometry('350x550')    #window size
root.configure(background = 'white')
root.resizable(0,0)

root.columnconfigure(0, pad=3)
root.columnconfigure(1, pad=3)
root.columnconfigure(2, pad=3)
root.columnconfigure(3, pad=3)

root.rowconfigure(0, pad=3)
root.rowconfigure(1, pad=3)
root.rowconfigure(2, pad=3)
root.rowconfigure(3, pad=3)
root.rowconfigure(4, pad=3)

#========================#
txt = StringVar()
a = float()
b = float()
op = int()
#========================#
def concat(a):
    z = str(txt.get())
    if len(z) < 10:
        txt.set(z + str(a))
    
    
def clear():
    txt.set('')
    
    
def op(x):
    global a, op
    if txt.get() == '':
        return
    else:
        a = float(txt.get())
        op = x
    clear()


def equal():
    global a, b, op
    b = str(txt.get())
    if op == 0 or b == '':
        return
    elif op == 1:
        clear()
        ans = a+float(b)
        if ans > 9999999999:
            txt.set('{:.2e}'.format(ans))
        else:
            txt.set('{:.2f}'.format(ans))
        op = 0
    elif op == 2:
        clear()
        ans = a-float(b)
        if ans > 9999999999:
            txt.set('{:.2e}'.format(ans))
        else:
            txt.set('{:.2f}'.format(ans))
        op = 0
    elif op == 3:
        clear()
        ans = a*float(b)
        if ans > 9999999999:
            txt.set('{:.2e}'.format(ans))
        else:
            txt.set('{:.2f}'.format(ans))
        op = 0
    elif op == 4:
        clear()
        ans = a/float(b)
        if ans > 9999999999:
            txt.set('{:.2e}'.format(ans))
        else:
            txt.set('{:.2f}'.format(ans))
        op = 0
        
        
#========================#
display = Label(root, textvariable = txt, font = "Arial 60 bold italic", bg = '#696969', fg = 'white', anchor = 'se', width = '10',
                height = '2', padx = '3', pady = '1.5')
txt.set('')

btn1 = Button(root, text = "1", font = "Arial 20 bold", width = 6, height = 3, command = partial(concat,1)) 
btn2 = Button(root, text = "2", font = "Arial 20 bold", width = 6, height = 3, command = partial(concat,2))
btn3 = Button(root, text = "3", font = "Arial 20 bold", width = 6, height = 3, command = partial(concat,3))
btn4 = Button(root, text = "4", font = "Arial 20 bold", width = 6, height = 3, command = partial(concat,4)) 
btn5 = Button(root, text = "5", font = "Arial 20 bold", width = 6, height = 3, command = partial(concat,5))
btn6 = Button(root, text = "6", font = "Arial 20 bold", width = 6, height = 3, command = partial(concat,6))
btn7 = Button(root, text = "7", font = "Arial 20 bold", width = 6, height = 3, command = partial(concat,7)) 
btn8 = Button(root, text = "8", font = "Arial 20 bold", width = 6, height = 3, command = partial(concat,8))
btn9 = Button(root, text = "9", font = "Arial 20 bold", width = 6, height = 3, command = partial(concat,9))
btn0 = Button(root, text = "0", font = "Arial 20 bold", width = 13, height = 3, command = partial(concat,0))
dec = Button(root, text = ".", font = "Arial 20 bold", width = 6, height = 3, command = partial(concat, '.'))

add = Button(root, text = "+", font = "Arial 20 bold", width = 6, height = 3, command = partial(op, 1)) 
sub = Button(root, text = "-", font = "Arial 20 bold", width = 6, height = 3, command = partial(op, 2))
mult = Button(root, text = "x", font = "Arial 20 bold", width = 6, height = 3, command = partial(op, 3))
div = Button(root, text = "/", font = "Arial 20 bold", width = 6, height = 3, command = partial(op, 4))
eq = Button(root, text = "=", font = "Arial 20 bold", width = 6, height = 3, command = equal)
clr = Button(root, text = "C", font = "Arial 20 bold", width = 6, height = 3, command = clear)

#========================#
display.grid(row = 0, columnspan=4, sticky=W+E)
btn1.grid(row = 4, column = 0)
btn2.grid(row = 4, column = 1)
btn3.grid(row = 4, column = 2)
btn4.grid(row = 3, column = 0)
btn5.grid(row = 3, column = 1)
btn6.grid(row = 3, column = 2)
btn7.grid(row = 2, column = 0)
btn8.grid(row = 2, column = 1)
btn9.grid(row = 2, column = 2)
btn0.grid(row = 5, columnspan = 2)
dec.grid(row = 5, column = 2)

div.grid(row = 4, column = 3)
mult.grid(row = 3, column = 3)
sub.grid(row = 2, column = 3)
add.grid(row = 1, column = 3)
eq.grid(row = 5, column = 3)
clr.grid(row = 1, column = 0)

root.mainloop()
```
